```python
from sklearn.cluster import KMeans
import pickle
class KMeansCluster():
    def __init__(self,x):
        self.x = x
        self.wcss=[]
    def cluster_no(self):
        for i in range(1,11):
            kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
            kmeans.fit(self.x)
            self.wcss.append(kmeans.inertia_)
        plt.plot(range(1,11),self.wcss)
        plt.xlabel("number of cluster")
        plt.ylabel("wcss")
        plt.show()
    def create_cluster(self,number_of_cluster):
        kmeans = KMeans(n_clusters=number_of_cluster,init='k-means++',random_state=42)
        self.x["cluster_no"] = kmeans.fit_predict(self.x)
        pickle.dump(kmeans, open("models/kmeans", 'wb'))
        return self.x
```

```python
X = df_temp.copy()
kmeans_object = KMeansCluster(X)
kmeans_object.cluster_no()
```

```python
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
def get_best_parameters_linearRegression(x_train,y_train):
    parameters = {"fit_intercept":[True,False],"normalize":[True,False],"copy_X":[True,False]}
    gride = GridSearchCV(LinearRegression(),parameters,cv=5,verbose=3)
    gride.fit(x_train,y_train)
    fit_intercept = gride.best_params_["fit_intercept"]
    normalize = gride.best_params_["normalize"]
    copy_x = gride.best_params_["copy_X"]
    parameter_linear = {"fit_intercept":fit_intercept,"normalize":normalize,"copy_X":copy_x}
    #mlflow.log_params("LinearRegression parameters",parameter_linear)

    Linear = LinearRegression(fit_intercept=fit_intercept,normalize=normalize,copy_X=copy_x)
    Linear.fit(x_train,y_train)
    #mlflow.sklearn.log_model(Linear,"Linear Regression")
    return Linear

def get_best_parameters_randomforest(x_train,y_train):
    parameters = {"n_estimators":[10,20,30],"min_samples_split":[2,4,8],
        "bootstrap":[True,False] 
}
    gride = GridSearchCV(RandomForestRegressor(),parameters,cv=5,verbose=3)
    gride.fit(x_train,y_train)
    n_estimators = gride.best_params_["n_estimators"]
    min_samples_split = gride.best_params_["min_samples_split"]
    bootstrap = gride.best_params_["bootstrap"]

    parameters_RandomForest = {"n_estimators":n_estimators,"min_samples_split":min_samples_split,"bootstrap":bootstrap}

    #mlflow.log_params("RandomForestRegressor parameters",parameters_RandomForest)

    random_forest = RandomForestRegressor(n_estimators=n_estimators,min_samples_split=min_samples_split,bootstrap=bootstrap)
    random_forest.fit(x_train,y_train)
    #mlflow.sklear.log_model(random_forest,"RandomForestRegressor")
    return random_forest

def get_best_model(x_train,x_test,y_train,y_test):

    linear_Regression = get_best_parameters_linearRegression(x_train,y_train)
    predicted_linear = linear_Regression.predict(x_test)
    linear_regression_r2_score = r2_score(y_test,predicted_linear)
    #mlflow.log_metric("linear Regression r2_score",r2_score(y_test,predicted_linear))


    Radom_forest = get_best_parameters_randomforest(x_train,y_train)
    predicted_forest = Radom_forest.predict(x_test)
    Random_forest_r2_score = r2_score(y_test,predicted_forest)


    if linear_regression_r2_score>Random_forest_r2_score:
        return linear_Regression,"Linear Regression"
    else:
        return Radom_forest , "Random forest Regression"
```

```python
import os
cluster_model = {}
for i in list_of_unique_clusters:
    cluster_df = df_new[df_new["cluster_no"]==i]
    cluster_x = cluster_df.drop(["cluster_no","labels"],axis=1)
    cluster_y = cluster_df["labels"]
    x_train,x_test,y_train,y_test = train_test_split(cluster_x,cluster_y,test_size=0.2,random_state=42)
    model , model_name = get_best_model(x_train,x_test,y_train,y_test)
    if os.path.isdir("models"):
        pass
    else:
        os.mkdir("models")
    print(f"For cluster {i} model {model_name}")
    cluster_model_path = "models/"+model_name+"_cluster"+str(i)
    cluster_model[i] =  cluster_model_path
    pickle.dump(model, open(cluster_model_path, 'wb'))
```
